```python
# Check dataset files
print(os.listdir(DATASET))
```

```python
# Read dataset information
yaml_path = DATASET + r"\data.yaml"

with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

print(data)
```

```python
# Count images
train = os.path.join(DATASET,"fruit-detection-dataset","images", "train")

val = os.path.join(DATASET,"fruit-detection-dataset","images","val")

print("Train exists:", os.path.exists(train))
print("Val exists:", os.path.exists(val))

print("Train images =", len(os.listdir(train)))
print("Validation images =", len(os.listdir(val)))
```

```python
# Show one image
from PIL import Image
import matplotlib.pyplot as plt

img_path = train + "\\" + os.listdir(train)[0]

img = Image.open(img_path)

plt.figure(figsize=(6,6))
plt.imshow(img)
plt.axis("off")

print(img_path)
```

```python
# Load pretrained YOLO model
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```
